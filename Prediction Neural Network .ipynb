```python
# Compute MAE
def compute_error(y_hat, y):
    # mean absolute error
    return np.abs(y_hat - y).mean()

def compute_error_normalized(y_hat, y):
    sum_peaks = [];
    for i in range(len(y)):
        if (i == 0 or y[i-1] < y[i]):
            if (i == len(y)-1 or y[i] > y[i+1]):
                sum_peaks.append(y[i])
                
    return np.abs(y_hat - y).mean() / np.array(sum_peaks).mean()
```
