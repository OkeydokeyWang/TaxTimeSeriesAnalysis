## Build data set

```python
data = np.hstack((
    feat1,
    feat2,
    feat3, 
    feat4,
    feat5,
    val1,
    val2))
```

```python
data.shape
```

```python
data = data.T
```

```python
# Split train validation,and test
train_data = data[:,:1000]
test_data = data[:,1000:]
```

```python
train_x = train_data[:-2,:].T
train_y = train_data[-2:,:].T

test_x = test_data[:-2,:].T
test_y = test_data[-2:,:].T
```

## DT w/ Cross Validation

```python
from sklearn import tree
from sklearn.model_selection import KFold
```

```python
# Compute MAE
def compute_error(y_hat, y):
    # mean absolute error
    return np.abs(y_hat - y).mean()
```

```python
CRITERIA = ["mse", "mae"]
MAX_DEPTHS = [6, 9, 12, 15, 19]
Ks = [4,5,7,9]
```

```python
# Vars to store results
cval_errs = {} # Mean validation errors
train_time = {} # Training time
# Best Model
best_af_dt_regressor = None
best_err = sys.maxsize


for criterion in CRITERIA:
    for max_dep in MAX_DEPTHS:
        for K in Ks:
            kf = KFold(n_splits=K, random_state=None, shuffle=False)
            y_err = []

            start = time.time()

            # Cross Validaiton
            for train_index, val_index in kf.split(train_x):
            #     print("TRAIN:", train_index, "VAL:", val_index)
                X_train, X_val = train_x[train_index], train_x[val_index]
                y_train, y_val = train_y[train_index], train_y[val_index]

                DTRegressor = tree.DecisionTreeRegressor(criterion=criterion, max_depth=max_dep)
                y_hat = DTRegressor.fit(X_train, y_train).predict(X_val)
            #     print(y_hat)
                y_err.append(compute_error(y_hat, y_val))

            end = time.time()

            print(str(criterion), str(max_dep), "mean val MAE:", np.mean(y_err))
#             print("Time lapsed", str((end - start)*1000))

            # add to dict
            cval_errs[str(criterion)+str(max_dep)] = np.mean(y_err)
            train_time[str(criterion)+str(max_dep)] = (end - start)*1000
            if np.mean(y_err) < best_err:
                best_af_dt_regressor = tree.DecisionTreeRegressor(criterion=criterion, max_depth=max_dep)
                best_af_dt_regressor = best_af_dt_regressor.fit(train_x, train_y)
                best_err = np.mean(y_err)
                best_errs = y_err
                
print(best_af_dt_regressor, best_err)
```

# Test
