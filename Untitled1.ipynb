# Load Feature 1 euro exchange rate from CSV

```python
df = pd.read_csv("./independendt projecct/euro_exchange_rate.csv")
```

```python
# df 前7行去掉 9.30.2019
df = df.iloc[7:,]
```

```python
df.head()
```

# Process Data

```python
import numpy as np
```

```python
# 取出exchange rate值
a=df.iloc[:,0:]
euro_byday=df.iloc[:,0:].values

print(type(df))
print(type(euro_byday))
```

```python
# 去除空值
for i in range(len(euro_byday)):
    if (euro_byday[i] == "-"):
        j = i
        while euro_byday[j] == "-":
            j = j - 1
        euro_byday[i] = euro_byday[j]
```

```python
euro_byday_float = euro_byday.astype('float')
```

```python
euro_byday_float = np.flip(euro_byday_float)
euro_byday_float
```

# Visualize

```python
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = (20,10)
```

```python
plt.plot(euro_byday_float)
```
